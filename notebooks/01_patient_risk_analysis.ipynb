```python
# ðŸ¥ Patient Risk Analysis Module
**Project:** Smart-Patient-Triage-System
**Module:** Data Processing & Risk Filtering

### ðŸŽ¯ Objective
Notebook ini mensimulasikan proses pengolahan data Rekam Medis Elektronik (EMR). Tujuannya adalah untuk mengidentifikasi pasien berisiko tinggi berdasarkan demografi (umur) dan kompleksitas gejala.

### âš™ï¸ Key Steps:
1.  **Data Ingestion:** Membuat data dummy pasien (simulasi database rumah sakit).
2.  **Feature Engineering:** Mengubah teks gejala mentah menjadi data terstruktur (tokenisasi).
3.  **Risk Filtering:** Menyaring pasien berdasarkan kriteria risiko medis tertentu (Usia > 40 & Jumlah Gejala > 2).
```

Import Libary

```python
import pandas as pd
import os

pd.set_option('display.max_columns', None)
print("Libraries loaded successfully.")
```

Data Mocking / Simulasi Data

```python
# 1. Simulasi Data Pasien (Mock Data)
# Dalam skenario nyata, data ini biasanya ditarik dari SQL Database atau API RS.
raw_data = [
    {'id': 1, 'name': 'Andi', 'age': 45, 'symptoms': 'demam, batuk, sesak napas'},
    {'id': 2, 'name': 'Budi', 'age': 29, 'symptoms': 'mual, sakit perut'},
    {'id': 3, 'name': 'Citra', 'age': 62, 'symptoms': 'pusing, kehilangan keseimbangan'},
    {'id': 4, 'name': 'Dita', 'age': 30, 'symptoms': 'susah tidur'},
    {'id': 5, 'name': 'Eka', 'age': 18, 'symptoms': 'gusi berdarah'},
    {'id': 6, 'name': 'Fitra', 'age': 49, 'symptoms': 'pusing, sakit perut'},
    {'id': 7, 'name': 'Gio', 'age': 49, 'symptoms': 'menggigil, batuk, sakit kepala'},
    {'id': 8, 'name': 'Harianto', 'age': 33, 'symptoms': 'memar di tangan'},
    {'id': 9, 'name': 'Idul', 'age': 88, 'symptoms': 'susah tidur'},
    {'id': 10, 'name': 'Jaka', 'age': 54, 'symptoms': 'sesak nafas'}
]

# Konversi ke Pandas DataFrame
df_patients = pd.DataFrame(raw_data)

# (Opsional) Simpan ke CSV sebagai raw file
csv_filename = 'raw_patients_data.csv'
df_patients.to_csv(csv_filename, index=False)

print(f"Data Loaded: {len(df_patients)} records.")
display(df_patients.head())
```

Feature Engineering

```python
# 2. Feature Engineering: Symptom Tokenization
# Tujuan: Mengubah string gejala (teks mentah) menjadi list agar bisa dihitung jumlahnya.

# Langkah A: Split string berdasarkan koma menjadi list
# Contoh: "demam, batuk" -> ["demam", "batuk"]
df_patients['symptoms_list'] = df_patients['symptoms'].astype(str).apply(lambda x: x.split(', '))

# Langkah B: Menghitung jumlah gejala (Symptom Count)
# Ini penting untuk menentukan tingkat urgensi pasien
df_patients['symptom_count'] = df_patients['symptoms_list'].apply(lambda x: len(x))

print("Feature Engineering complete. New columns added: 'symptoms_list', 'symptom_count'")
display(df_patients[['name', 'symptoms', 'symptom_count']].head())
```

Filtering

```python
# 3. Risk Identification Logic
# Kita mendefinisikan "High Risk Patient" dengan aturan bisnis berikut:
# - Aturan 1: Usia di atas 40 tahun (Faktor risiko degeneratif)
# - Aturan 2: Memiliki lebih dari 2 gejala sekaligus (Indikasi komplikasi)

RISK_AGE_THRESHOLD = 40
RISK_SYMPTOM_THRESHOLD = 2

# Menerapkan Filter
filter_age = df_patients['age'] > RISK_AGE_THRESHOLD
filter_symptoms = df_patients['symptom_count'] > RISK_SYMPTOM_THRESHOLD

high_risk_patients = df_patients[filter_age & filter_symptoms].copy()

print(f"Filter Criteria: Age > {RISK_AGE_THRESHOLD} AND Symptoms > {RISK_SYMPTOM_THRESHOLD}")
print(f"High Risk Patients Found: {len(high_risk_patients)}")
```

Result & Insight

```python
# 4. Final Result
# Menampilkan data pasien yang butuh penanganan prioritas
print("=== HIGH RISK PATIENT LIST ===")
display(high_risk_patients)

# (Opsional) Export hasil filter untuk tim medis
high_risk_patients.to_csv('high_risk_patients_report.csv', index=False)
print("Report saved successfully.")
```
